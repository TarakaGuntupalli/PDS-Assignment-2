# a) set a seed (to ensure work reproducibility) and take a random sample of 25 observations and find the mean Glucose and highest Glucose values of this sample and compare these statistics with the population statistics of the same variable. You should use charts for this comparison.

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# load the data
diabetes = pd.read_csv('diabetes.csv')

# set the seed
np.random.seed(21)

# sample of 25 random observations
sample = diabetes.sample(n=25)

# mean and max glucose values of sample
sample_mean_glucose = sample['Glucose'].mean()
sample_max_glucose = sample['Glucose'].max()


print("Sample mean glucose:", sample_mean_glucose)
print("Sample max glucose:", sample_max_glucose)

# mean and max glucose values of population
population_mean_glucose = diabetes['Glucose'].mean()
population_max_glucose = diabetes['Glucose'].max()

print("Population mean glucose:", population_mean_glucose)
print("Population max glucose:", population_max_glucose)

# plot the comparison
plt.bar(['Sample Mean', 'Population Mean'], [sample_mean_glucose, population_mean_glucose])
plt.title('Comparison of Mean Glucose Values')
plt.show()

plt.bar(['Sample Max', 'Population Max'], [sample_max_glucose, population_max_glucose])
plt.title('Comparison of Max Glucose Values')
plt.show()
```

# b) Find the 98th percentile of BMI of your sample and the population and compare the results using charts.

```python
# 98th percentile of BMI of the sample and the population
Sample_98th_percentile = np.percentile(sample['BMI'], 98)
population_98th_percentile = np.percentile(diabetes['BMI'], 98)

print("Sample 98th percentile:", Sample_98th_percentile)
print("Population 98th percentile:", population_98th_percentile)

# plot the comparison
plt.bar(['Sample 98th Percentile', 'Population 98th Percentile'], [Sample_98th_percentile, population_98th_percentile])
plt.title('Comparison of 98th Percentile of BMI')
plt.show()
```

# c) Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the population and find the average mean, standard deviation and percentile for BloodPressure and compare this with these statistics from the population for the same variable. Again, you should create charts for this comparison. Report on your findings.

```python
# set the seed
np.random.seed(21)

#  500 samples of 150 observation each from the population
n_samples = 500
sample_size = 150
samples = np.random.choice(diabetes['BloodPressure'], size=(n_samples, sample_size), replace=True)

# mean, standard deviation, and percentile of BloodPressure of each sample
means = np.mean(samples, axis=1)
stds = np.std(samples, axis=1)
percentiles = np.percentile(samples, q=95, axis=1)

# mean, standard deviation, and percentile of BloodPressure of the population
population_mean = diabetes['BloodPressure'].mean()
population_std = diabetes['BloodPressure'].std()
population_percentile = np.percentile(diabetes['BloodPressure'], q=95)

# plot the comparison of means
plt.hist(means, alpha=0.5, label='Sample Means')
plt.axvline(x=population_mean, color='red', label='Population Mean')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Mean BloodPressure')
plt.show()

# plot the comparison of standard deviations
plt.hist(stds, alpha=0.5, label='Sample Stds')
plt.axvline(x=population_std, color='red', label='Population Std')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Std BloodPressure')
plt.show()

# plot the comparison of percentiles
plt.hist(percentiles, alpha=0.5, label='Sample Percentiles')
plt.axvline(x=population_percentile, color='red', label='Population Percentile')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Percentile BloodPressure')
plt.show()
```
